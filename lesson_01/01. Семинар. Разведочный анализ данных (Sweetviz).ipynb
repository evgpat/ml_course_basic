```python
import sweetviz as sv
import pandas as pd

from sklearn.datasets import load_wine # для загрузки датасета
```

```python
# загружаем датасет
data = load_wine(as_frame=True)

# извлекаем данные и сохраняем их как pandas.DataFrame
df = data.frame
```

```python
# формируем отчет
report = sv.analyze(df)
report.show_html('sweetviz_wine.html') # файл сохранится в виде html-документа, который нужно скачать и открыть отдельно
```
